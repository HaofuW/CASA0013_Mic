## 1. Load the data

### 1.1 London Borough Map

### 1.2 Inside Airbnb

```python
# load the data and select the columns needed
cols = ['id','name','price','neighbourhood_cleansed','latitude','longitude','property_type','room_type']
url_airbnb = 'http://data.insideairbnb.com/united-kingdom/england/london/2021-10-10/data/listings.csv.gz'
df_airbnb = pd.read_csv(url_airbnb, compression='gzip', usecols=cols)

df_airbnb.info()
```

#### Data Cleaning

```python
# clean the data
df_airbnb['price'] = df_airbnb.price.str.replace('$','').str.replace(',','').astype(float)
df_airbnb.drop(df_airbnb[((df_airbnb.latitude.isna())|(df_airbnb.longitude.isna()))].index, axis=0, inplace=True)
df_airbnb.drop(df_airbnb[((df_airbnb.latitude < 40)|(df_airbnb.longitude > 1))].index, axis=0, inplace=True)

df_airbnb.info
```

```python
# focus on "Private room" in the room type
ab_room_type = 'Private room'
df_airbnb = df_airbnb[df_airbnb.room_type == ab_room_type]
```

### 1.3 Rental

```python
# load the data and select the columns needed
cols_rental = ['Borough','Bedroom_Category','Count_of_rents','Mean']
url_rental = 'https://github.com/HaofuW/CASA0013_Mic/blob/main/Assessment3/Code/london_rental_statistics.xlsx?raw=true'
df_rental = pd.read_excel(url_rental, sheet_name = 0, usecols = cols_rental)

df_rental.info()
```

#### Data Cleaning

```python
# clean the data
df_rental.drop(index = df_rental[df_rental.Count_of_rents.isna()].index, axis=0, inplace=True)
df_rental.drop(index = df_rental[df_rental.Mean.isna()].index, axis = 0, inplace=True)

print(df_rental.shape)
df_rental.sample(3, random_state=42)
```

```python
# focus on "Room" in the rental market
rental_room_type = 'Room'
df_rental = df_rental[df_rental.Bedroom_Category == rental_room_type]
```

## 3. Mapping

### 3.1 Private Room of Airbnb Distributions

#### Add Geometry and Reproject

```python
# Convert the df into a GeoDataFrame named gdf
gdf_airbnb = gpd.GeoDataFrame(df_airbnb, geometry=gpd.points_from_xy(df_airbnb.longitude, df_airbnb.latitude, crs='epsg:4326'))
gdf_airbnb = gdf_airbnb.to_crs('epsg:27700')
```

#### Add LA Names to Data

```python
# Associate LA (Local Authority) names to the listings using a spatial join
gdf_boros = gpd.sjoin(gdf_airbnb, boros, op='within', how='left')
print(gdf_boros.columns.values)
```

```python
print(gdf_boros.shape)
gdf_boros.sample(3, random_state=42)[['id','name','NAME']]
```

#### Tidy Up

```python
gdf_boros.drop(columns=['index_right','HECTARES','NONLD_AREA','ONS_INNER'], inplace=True)
gdf_boros.NAME.unique()
```

```python
gdf_airbnb[gdf_boros.NAME.isna()][['id','name']]
```

```python
# drop problematic results
gdf_boros.drop(index=gdf_airbnb[gdf_boros.NAME.isna()].index, axis=1, inplace=True)
print(gdf_boros.shape)
```

#### Check Results

```python
# plot the locations of airbnb listings
def map_airbnb_loc():
    
    ax = gdf_boros.plot(column='NAME', markersize=0.5, alpha=0.5, figsize=(16,10))
    boros.plot(ax=ax, edgecolor='0.5', facecolor='None', alpha=0.5);
    ax.axis('off')

    ax.set_title('Private Room Location Distributions of Airbnb listings in London Boroughs', fontdict={
             'fontsize': '16', 'fontweight': '1'})
    
map_airbnb_loc()
```

## 4. OLS Regression

### 4.1 Average Price of Single Room in Rental Market & Counts of Private Room in Airbnb Listings

#### Linear Regression

```python
# set the average price of airbnb as x value
# set the average price of rental market as y value
x_values = boros_rental_airbnb_count['count']
y_values = boros_rental_airbnb_count['Mean']
```

```python
x_values
```

```python
y_values
```

```python
import statsmodels.api as sms

X_values = sms.add_constant(x_values)
regression_model_a = sms.OLS(y_values, X_values)
regression_model_b = regression_model_a.fit()

print(regression_model_b.summary())
print()
```

```python
p_value    = regression_model_b.f_pvalue
R_squared  = regression_model_b.rsquared
print("R_squared  =", R_squared)
print("p_value    =", p_value)
```

#### Linear Regression after Logarithm

```python
import statsmodels.formula.api as sm

regression_model_log = sm.ols(formula="Mean ~ count.apply(np.log)", data = boros_rental_airbnb_count).fit()
print(regression_model_log.params)
```

```python
print(regression_model_log.summary())
```

```python
p_value_log    = regression_model_log.f_pvalue
R_squared_log  = regression_model_log.rsquared
print("R_squared_log  =", R_squared_log)
print("p_value_log    =", p_value_log)
```

## Background


With the rapid development of the sharing economy, the sharing platforms such as Airbnb have hugely affected the traditional economy market (Muschter, Caldicott et al. 2020). The dominant advantage of the Airbnb is the peer-to-peer short-term accommodation platforms where the residents have a chance to provide their own properties for the people with demand such as tourists (Li, Kim et al. 2019). However, due to the lower price, less regulations and online booking, more and more people tend to choose the Airbnb for long-term accommodation instead of the traditional rental companies (Li, Kim et al. 2019). 

As a result, to prevent the further unregulated scenarios, Airbnb released the 90-day rule for the listings in London that customers are not allowed to take any bookings if the night limits have reached 90 days unless the host has the planning permission of the property (Figueras, 2020). However, the rule doesnt include the single private room booking which according to the assumption, has a large impact of the traditional local rental markets. Here, for the further demonstration below, a Null Hypothesis that the rental market doesnt have any relationship with the Airbnb listings in the single private room type.


## Data Analysis

### Methodology

To tackle with the research hypothesis, the data cleaning and merging processes were conducted in both datasets of the Airbnb listings and the rental market. Apart from the cleanings of NAs in the datasets, the critical process was to pick out the single private room type from the two datasets.

The first method was the empirical analysis based on the distribution mappings of the counts of the Airbnb listings in private room in each borough and the average price of the rental market in single room. The empirical analysis here was used to conduct a rough relationship prediction of the Airbnb listings and the rental market.

To solve the inaccuracy caused by the first method, the OLS fitting was introduced to try to fit the dataset in a linear regression model. The variables taken here were the same with the first method, the counts of the private room in Airbnb listings in each borough and the average price of the single room in rental market. For the completeness of the research, the average price of the Airbnb private room was also taken to fit a regression model with the rental market. To achieve the best regression model, some of the variables might be calculated under the logarithm and the outliers could be detected and erased.


### Results

Figure 2, 3, 4 separately presents the several mappings containing the distribution of the counts of the private room in the Airbnb listings, the average price of the Airbnbs private room by borough and the average price of the single room in the rental market by borough. What needs to be mentioned here is that the blank area in Figure 4 indicates that the borough has less than 5 observations which could be further discussed in the Conclusion.
